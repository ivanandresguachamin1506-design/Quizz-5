```python
import random
class Jugador:
    def __init__(self, nombre, monedas):
        self.nombre = nombre
        self.monedas = monedas
    def apostar(self):
        self.monedas -= 10
    def recibir_premio(self, premio):
        self.monedas += premio
    def mostrar_saldo(self):
        print(f"Monedas: {self.monedas}")
```

```python
print("Bienvenido a la mejor tragaperras de New VEGAS 🎰")
print("Cada tirada cuesta 10 chapas")
simbolos = ["💸", "𝟕𝟕𝟕","🍀","🍒", "💎","💰","💵"]
jugador = Jugador("Iván", 100)
while jugador.monedas > 0:
    jugador.mostrar_saldo()
    opcion = input("¿Deseas tirar? (s/n): ")
    if opcion.lower() != "s":
        break
        jugador.apostar()
    tirada = [random.choice(simbolos) for _ in range(3)]
    print(tirada)
    if tirada[0] == tirada[1] == tirada[2]:
        print("¡3 iguales! Ganas 30 monedas")
        jugador.recibir_premio(-10)
        jugador.recibir_premio(30)
    elif tirada[0] == tirada[1] or tirada[1] == tirada[2] or tirada[0] == tirada[2]:
        print("¡2 iguales! Ganas 15 monedas")
        jugador.recibir_premio(15)
        jugador.recibir_premio(-10)
    else:
        print("No ganaste nada")
        jugador.recibir_premio(-10)

print("Juego terminado \nRecuerda que la casa nunca pierde")
  
```
